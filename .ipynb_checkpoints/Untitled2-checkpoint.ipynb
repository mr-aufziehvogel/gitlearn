```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv("survey_results_public.csv")
```

```python
df.columns
```

```python
countries = df.Country.value_counts()
```

```python
countries.keys()
```

```python
plt.pie(countries, labels=countries.keys())
plt.tight_layout()
```

```python
print(*countries)
```

```python
df2 = pd.DataFrame(data=countries)
df2.rename(columns={"Country":"Users"}, inplace=True)
```

```python
df3 = df2[df2["Users"] > 500]
```

```python
df3.index.values
```

```python
plt.pie(df3.Users, labels=df3.index.values)
plt.tight_layout()
```
