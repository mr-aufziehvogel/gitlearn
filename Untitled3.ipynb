```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("results.csv", encoding = 'unicode_escape', usecols=[0,1,2,3,4,5,6])
```

```python
df = df[df["Season"] == "2019-20"]
```

```python
def winner(row):
    if row["FTR"] == "H":
        return row["HomeTeam"]
    elif row["FTR"] == "A":
        return row["AwayTeam"]
    else:
        return "Draw"

df['Winner'] = df.apply(winner, axis=1)

def points(row):
    if row["FTR"] == "H":
        return 3
    elif row["FTR"] == "A":
        return 0
    else:
        return 1

df['Points'] = df.apply(points, axis=1)

def pointsaway(row):
    if row["FTR"] == "H":
        return 0
    elif row["FTR"] == "A":
        return 3
    else:
        return 1
    
    
df['PointsAway'] = df.apply(pointsaway, axis=1)
#htlist = list(df["HomeTeam"])
#atlist = list(df["AwayTeam"])
#ftrlist = list(df["FTR"])

#winners = list(map(lambda x,y,z: winner(x,y,z),htlist,atlist,ftrlist))




#df["Winner"]

#df["Winner"].apply(lambda x: winner(ht = df["HomeTeam"], at = df["AwayTeam"], ftr = df["FTR"]))
    
#df["Winner"].map((lambda x,y,z = winner(x,y,z),df.HomeTeam,df.AwayTeam,df.FTR))
```

```python
p_home = df.groupby('HomeTeam').sum()['Points'].sort_values(ascending=False)
```

```python
p_away = df.groupby('AwayTeam').sum()['PointsAway'].sort_values(ascending=False)
```

```python
table = p_home + p_away
```

```python
table = pd.DataFrame(table.sort_values(ascending = False))
```

```python
table
```

```python
table.columns = ["Points"]
table.index.name = "Team"

marktwert = [1.09, 1.27, 0.775, 0.797, 0.384, 1.04, 0.315, 0.680, 0.205, 0.110, 0.259, 0.525, 0.312, 0.248, 0.225, 0.371, 0.184, 0.311, 0.270, 0.110]
```

```python
table
```

```python
table["Market value"] = marktwert
```

```python
np_mk = table["Market value"].to_numpy().reshape((-1, 1))
np_points = table["Points"].to_numpy()
```

```python
model = LinearRegression().fit(np_mk, np_points)
regr = model.predict(np_mk)
```
