```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("results.csv", encoding = 'unicode_escape', usecols=["DateTime", "FTHG", "FTAG"])
```

```python
df
```

```python
def winner(row):
    if row["FTR"] == "H":
        return row["HomeTeam"]
    elif row["FTR"] == "A":
        return row["AwayTeam"]
    else:
        return "Draw"

df['Winner'] = df.apply(winner, axis=1)

def points(row):
    if row["FTR"] == "H":
        return 3
    elif row["FTR"] == "A":
        return 0
    else:
        return 1

df['Points'] = df.apply(points, axis=1)

def pointsaway(row):
    if row["FTR"] == "H":
        return 0
    elif row["FTR"] == "A":
        return 3
    else:
        return 1
    
    
df['PointsAway'] = df.apply(pointsaway, axis=1)
#htlist = list(df["HomeTeam"])
#atlist = list(df["AwayTeam"])
#ftrlist = list(df["FTR"])

#winners = list(map(lambda x,y,z: winner(x,y,z),htlist,atlist,ftrlist))




#df["Winner"]

#df["Winner"].apply(lambda x: winner(ht = df["HomeTeam"], at = df["AwayTeam"], ftr = df["FTR"]))
    
#df["Winner"].map((lambda x,y,z = winner(x,y,z),df.HomeTeam,df.AwayTeam,df.FTR))
```

```python
p_home = df.groupby('HomeTeam').sum()['Points'].sort_values(ascending=False)
```

```python
p_away = df.groupby('AwayTeam').sum()['PointsAway'].sort_values(ascending=False)
```

```python
g_home = df.groupby('HomeTeam').sum()['FTHG'].sort_values(ascending=False)
```

```python
g_away = df.groupby('AwayTeam').sum()['FTAG'].sort_values(ascending=False)
```

```python
g_away
```

```python
table = p_home + p_away
```

```python
table = pd.DataFrame(table.sort_values(ascending = False))
```

```python
table["HG"] = g_home
table["AG"] = g_away
table["Diff"] = (g_home - g_away) / (g_home + g_away)
table.sort_values(by="Diff", ascending=False)
```
